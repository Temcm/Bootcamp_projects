```python
import requests
import time
import pandas as pd
```

```python
names = [ ] 
heights = [ ] 
masses = [ ] 

for i in range(1,11):
    url = f"https://swapi.dev/api/people/{i}"
    resp = requests.get(url)
    result = resp.json()
    names.append (result ['name'] )
    heights.append (result ['height'] )
    masses.append (result ['mass'] )
    time.sleep(1)

df = pd.DataFrame({ 
        "name": names, "height": heights, "mass": masses

})

df
```
